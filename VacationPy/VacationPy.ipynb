### Store Part I results into DataFrame

```python
# File to Load
city_data_load = "../output_data/city_data.csv"

# Read City Data File and store into Pandas DataFrame
city_data = pd.read_csv(city_data_load)

city_data
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
locations = city_data[['Lat', 'Lng']]
humidity = city_data['Humidity']
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights = humidity,
                              dissipating=False)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
city_humidity = city_data[city_data.Humidity > 80]

city_wind = city_humidity.rename(columns={'Wind Speed':'Wind'})
city_wind = city_wind[city_wind.Wind < 10]
city_wind = city_wind[city_wind.Wind > 5]

city_cloud = city_wind[city_wind.Cloudiness > 50]

city_clean = city_cloud.rename(columns={'Wind':'Wind Speed'})
city_clean
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = city_clean.reset_index()
hotel_df
#print(len(hotel_df['City']))
```

```python
hotels = []
# geocoordinates
for i in range(0,len(hotel_df['City'])):
    target_coordinates = f"{hotel_df['Lat'][i]}, {hotel_df['Lng'][i]}"
    target_radius = 5000
    target_type = "hotel"

# set up a parameters dictionary
    params = {
        "location": target_coordinates,
        "radius": target_radius,
        "type": target_type,
        "key": g_key
    }

# base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
    response = requests.get(base_url, params=params).json()
#    print(target_coordinates)
#    pprint(response)

# Record Restaurant       
    try: 
        hotels.append(response["results"][1]["name"])
    
    except IndexError:
        hotels.append(np.nan)
```

```python
hotel_df['Hotel Name'] = hotels
hotel_clean = hotel_df.dropna()
hotel_df = hotel_clean.reset_index()
hotel_df
```
